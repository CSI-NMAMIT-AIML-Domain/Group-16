```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df_sales = pd.read_csv('Advertising And Sales.csv')  # Replace with your file path
top_10 = df_sales.nlargest(10, 'Sales')
# Bar plot for TV advertising budget vs Sales
plt.figure(figsize=(8, 6))
sns.barplot(x='TV', y='Sales', data=top_10)
plt.title('Sales vs TV Advertising Budget')
plt.show()

# Bar plot for Radio advertising budget vs Sales
plt.figure(figsize=(8, 6))
sns.barplot(x='Radio', y='Sales', data=top_10)
plt.title('Sales vs Radio Advertising Budget')
plt.show()

# Bar plot for Newspaper advertising budget vs Sales
plt.figure(figsize=(8, 6))
sns.barplot(x='Newspaper', y='Sales', data=top_10)
plt.title('Sales vs Newspaper Advertising Budget')
plt.show()

```
